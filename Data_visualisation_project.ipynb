```python
#Complaint type Breakdown with bar plot to figure out majority of complaint types and top 5 complaints
Seller_name = df.groupby(['Seller Name', 'Complaint Type']).size().unstack()
Seller_name.plot(kind='bar',figsize=(17,6));
```

```python
majorcomplints=df.dropna(subset=["Complaint Type"])
majorcomplints=df.groupby("Complaint Type")

sortedComplaintType = majorcomplints.size().sort_values(ascending=False)
sortedComplaintType = sortedComplaintType.to_frame('count').reset_index()

sortedComplaintType
sortedComplaintType.head()
```

```python
groupBycomplaintType = df.groupby('Complaint Type')
grp_data = groupBycomplaintType.get_group('Pickup not done')
grp_data.shape
```
